```python
#-*- coding: utf-8 -*-
import pandas as pd

'''
____________________________________________________________________________________
这里路径改一下就行，不需要就注释掉
————————————————————————————————————————————————————————————————————————————————————
'''
datafile= r'C:\Users\DELL\Desktop\data_cleaned.csv'  # 航空原始数据,第一行为属性标签
# 读取原始数据，指定UTF-8编码（需要用文本编辑器将数据装换为UTF-8编码）
data = pd.read_csv(datafile, encoding = 'utf-8')
data.head()
```

```python
# 构造属性L入会时长 连续变量 12.23~114.63
L = pd.to_datetime(data['LOAD_TIME']) - \
pd.to_datetime(data['FFP_DATE'])
L = L.astype('str').str.split().str[0]
L = L.astype('int')/30
data = data.copy()
data['L'] = L
data.head()
```

```python
#result为1即为流失客户，为0为非流失客户 
#对流失客户作出如下定义：最后一次乘机时间至观测窗口末端时长大于等于观测窗口内最大乘机间隔
data = data.copy()
data['result'] = 0
data = data.copy()
#data['result'][data['LAST_TO_END']>=data['MAX_INTERVAL']] = 1
data.loc[data['LAST_TO_END']>=data['MAX_INTERVAL'], 'result'] = 1
data.head()
```

```python
# 去除age空值
print(data.shape)
data = data[data['AGE'].notnull()]
data.shape
```

```python
# 将GENDER列进行硬编码
data.loc[data['GENDER']=='男', 'GENDER']=0
data.loc[data['GENDER']=='女', 'GENDER']=1
# 提取28个指标和对应是否流失result
data_new = data[['L','GENDER','FFP_TIER','AGE','FLIGHT_COUNT','SUM_YR_1','SUM_YR_2','SEG_KM_SUM','WEIGHTED_SEG_KM','AVG_FLIGHT_COUNT','AVG_BP_SUM','BEGIN_TO_FIRST','AVG_INTERVAL','L1Y_Flight_Count','P1Y_Flight_Count','avg_discount','Ration_P1Y_Flight_Count','Ration_L1Y_Flight_Count','P1Y_BP_SUM','L1Y_BP_SUM','EP_SUM','ADD_Point_SUM','Eli_Add_Point_Sum','Points_Sum','Point_NotFlight','ADD_POINTS_SUM_YR_1','ADD_POINTS_SUM_YR_2','EXCHANGE_COUNT','result']]
data_new
```
